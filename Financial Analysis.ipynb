```python
import pandas as pd
df =pd.read_csv('Finance_data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_style('darkgrid')
```

```python
df.head()
```

```python
df.drop(['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','Gold','PPF'],axis =1 , inplace=True)
```

```python
df=df.replace(r'\s*$',np.nan,regex=True)
```

```python
sns.countplot(x='age',data=df,palette="Set2",edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(data = df,  x="gender",hue="Investment_Avenues",palette='summer',linewidth=3,edgecolor="black")
plt.title("Investment Avenues")
plt.subplot(1,2,2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor="white")
plt.title("Stock Marktet")
```
